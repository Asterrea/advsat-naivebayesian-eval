The features that will be used are the words that can be found in the each document in the entire corpus. 

The code below, extracts the words given a filepath

***words =  [i for i in words if len(i) >= 3]*** filters put the words that are less than 3 letters


*** words = list(set(words))*** Creates a set from the list of words extracted, and duplicate words are removed



```python
import re
import os

def extractWords(filepath):
    file = open(filepath, 'r')
    # .lower() returns a version with all upper case characters replaced with lower case characters.
    text = file.read().lower()
    file.close()
    # replaces anything that is not a lowercase letter, a space, or an apostrophe with a space:
    text = re.sub('[^a-z]+', " ", text)
    words = list(text.split())
    
    words =  [i for i in words if len(i) >= 3] 
    words = list(set(words))
    
    return words;
```

The code below traverses all of the documents that can is under the directory

```python
import os

tokenList = []
rootdir = "Emails\\bare"
fileCtr = 0
# pattern = re.compile("\d+-\d+msg\d+.txt")
for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        tempList = []
        #print os.path.join(subdir, file)
        filepath =  subdir + os.sep + file
    
        tempList = extractWords(filepath)
        tokenList.extend(tempList)
        fileCtr = fileCtr + 1
            
tokenList = list(set(tokenList))     
print (tokenList)
```
